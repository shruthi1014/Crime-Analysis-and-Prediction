```python
df1=pd.read_csv("01_District_wise_crimes_committed_IPC_2013.csv")
df1.head()
```

```python
to_drop=['IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES','CUSTODIAL RAPE','COUNTERFIETING','CULPABLE HOMICIDE NOT AMOUNTING TO MURDER','PREPARATION AND ASSEMBLY FOR DACOITY','CRIMINAL BREACH OF TRUST','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY']
df1.drop(to_drop,inplace=True,axis=1)
df1.head()
```

```python
df1=df1[df1.DISTRICT!='TOTAL']
df1.head()
```

```python
datamerge=pd.merge(df,df1,how='outer')
datamerge.head()
```

```python
datamerge.rename(columns={'STATE/UT':'STATE'},inplace=True)
datamerge.rename(columns={'OTHER IPC CRIMES':'OTHERS','TOTAL IPC CRIMES':'TOTAL'},inplace=True)
to_drop=['AUTO THEFT','ARSON']
datamerge.drop(to_drop,inplace=True,axis=1)
datamerge=datamerge[datamerge.DISTRICT!='ZZ TOTAL']
datamerge.head()
```

```python
crime=input('Enter crime: ')
crime=crime.upper()
print(crime)
state=input('Enter state: ')
state=state.upper()
print(state)
yn=input('Do you want to enter district: ')
yn=yn.lower()
if(yn=='y'):
    district=input('Enter district: ')
    district=district.upper()
    print(district)
    ans=datamerge[datamerge.DISTRICT==district]
    show=ans.filter(items=['YEAR',crime]).groupby('YEAR').sum()
elif(yn=='n'):
    ans=datamerge[datamerge.STATE==state]
    show=ans.filter(items=['YEAR',crime])
    show=show.groupby('YEAR').sum()
```

```python
#Finding major crimes in a district
district=input('Enter district: ')
district=district.upper()
ans=datamerge[datamerge.DISTRICT==district]
ans.drop(['TOTAL','OTHERS','YEAR'],inplace=True,axis=1)
show=ans.groupby('DISTRICT').sum()
maximum_crime=show.idxmax(axis=1)
print("The most freuently occured crime is: ",maximum_crime[district])
show=show.loc[district]
show.plot(kind='pie')
```

```python
crime=input('Enter the crime: ')
ans=datamerge.filter(items=['DISTRICT',crime]).groupby('DISTRICT').sum()
ans.rename(columns={crime:'CRIME'},inplace=True)
ans=ans[ans.CRIME==ans.CRIME.max()]
print("The district with maximum ",crime," is: ")
print(ans)
```

```python
#Finding trend of total crimes in a district
district=input('Enter district to find trend: ')
ans=datamerge[datamerge.DISTRICT==district].filter(items=['YEAR','TOTAL']).groupby('YEAR').sum()
ans.plot(legend=True)
```

```python
crime=input('Enter crime: ')
year=int(input('Enter year(2003-2012): '))
ans=datamerge[datamerge.YEAR==year].filter(items=['STATE',crime]).groupby('STATE').sum()
ans=ans[crime]
ans.plot(kind='pie')
```

```python
from sklearn.linear_model import LinearRegression
district=input('Enter district to find trend: ')
ans=datamerge[datamerge.DISTRICT==district].filter(items=['YEAR','TOTAL']).groupby('YEAR').sum()
show=ans.reset_index()
x=show.iloc[:,0].values.reshape(-1,1)
y=show.iloc[:,1].values.reshape(-1,1)
linear_regressor=LinearRegression()
linear_regressor.fit(x,y)
y_pred=linear_regressor.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.show()
```

```python
def calc(slope,intercept,X):
    return slope*X+intercept
X=int(input("Enter year for prediction: "))
score=calc(linear_regressor.coef_, linear_regressor.intercept_, X)
Y=linear_regressor.predict([[X]])
print("The total crimes: ",Y)
```
